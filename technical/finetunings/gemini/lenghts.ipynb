```python
src = "gs://finetune-data123/test.src.txt"
summaries = "gs://finetune-data123/test.tgt.txt"
```

```python
from datasets import load_dataset
import pandas as pd

# Load datasets (line by line)
test_source = load_dataset("text", data_files=src)["train"]
test_summaries = load_dataset("text", data_files=summaries)["train"]

def length_stats(dataset):
    # Calculate lengths in characters and words
    char_lengths = [len(x["text"]) for x in dataset]
    word_lengths = [len(x["text"].split()) for x in dataset]

    stats = {
        "char_mean": sum(char_lengths) / len(char_lengths),
        "char_min": min(char_lengths),
        "char_max": max(char_lengths),
        "word_mean": sum(word_lengths) / len(word_lengths),
        "word_min": min(word_lengths),
        "word_max": max(word_lengths),
    }
    return stats

source_stats = length_stats(test_source)
summary_stats = length_stats(test_summaries)

# Combine into DataFrame for readability
results_df = pd.DataFrame([source_stats, summary_stats], index=["Source", "Summary"])
print(results_df)
```
